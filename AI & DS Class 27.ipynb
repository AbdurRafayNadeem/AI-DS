```python
import pandas as pd
import numpy as np
```

```python
students = {'name':['Ali','Faraz','Kamran','Faizan','Nasir','Asad','Danish'],
            'Age':[20,21,24,21,20,23,np.nan],
           'Python':[89,49,78,96,100,78,80],
           'Numpy':[78,87,np.nan,89,75,95,85],
           'Pandas':[40,75,85,48,60,89,96]}
data = pd.DataFrame(students)
data
```

```python
data['Total_Obt'] = data['Python'] + data['Numpy']+data['Pandas']
data
```

```python
data['Numpy'] = data['Numpy'].fillna(0)
```

```python
data
```

```python
data['Total_Obt'] = data['Python'] + data['Numpy']+data['Pandas']
data
```

```python
#To fill/drop NaN values you can use .fillna or .dropna
# data.fillna(0,inPlace=True) <-- yeh orginal mai change kardeta hai agar inPlace=True ho to
```

```python
data['Result'] = ["Pass" if m>=150 else "Try Again" for m in data['Total_Obt']]
```

```python
data
```

```python
data['Result'] = ["Pass" if (Py>=50 and Np>=50 and Pd>=50) else "Try Again" for Py, Np, Pd in zip(data["Python"],data["Numpy"],data["Pandas"])]
```

```python
data
```

```python
data['Percentage'] = round((data['Total_Obt']/300)*100,2)
```

```python
data
```

```python
data['Grade'] = ["A" if Per>=80 else "B" if Per>=70 "C" else "D" if Per>=60 "E" else "F" for Per in(data["Percentage"])]
```
